## Load the data

````python
def load_file(filename):
    import pandas as pd
    # Load the dataset
    data = pd.read_csv(filename)
    
    # Sample 5 rows from the dataset for prompt inclusion
    sample_data = data.sample(n=100, random_state=42)
    
    # Display the sample in a format suitable for the prompt
    sample_content = sample_data.to_markdown(index=False)
    return sample_content
````

## Define Prompt for Text-to-Code generation



````python
PROMPT_code = """
You are a software engineer.
Your task is to generate Python code to build a classification model based on the provided dataset.

You will be given a context paragraph describing the business problem and requirements. You will also be provided with a sample of the dataset structure to inform your understanding of the data.

Here are the steps to follow:

1. Read through the entire paragraph with the business context and dataset sample to understand the task:
<content>
{INPUT_CONTENT}
</content>

2. Read through the function description to understand the function you're going to write:
<function description>
{FUNCTION_DESCRIPTION}
</function description>

3. Generate the Python code according to the function description and the content provided. Please note that:
 a. Only implement the essential steps for building the model.
 b. Code should include data preprocessing, model training, and evaluation steps.
 c. **Use StandardScaler to scale the features before training**.
 d. **Set `max_iter` in LogisticRegression to 500** to avoid convergence issues.
 e. Provide only the code, without analysis or commentary.

4. Provide the final function based on the task requirements.
"""

````

````python
FUNCTION_DESCRIPTION = """
This function trains a classification model to predict whether a customer will renew their vehicle insurance based on the provided dataset.

Function name: train_insurance_renewal_model

Input variables:
- dataset : The dataset provided is from an automobile insurance business. The objective is to predict whether a customer will renew their vehicle insurance. The target column is Response, where 1 indicates renewal and 0 indicates no renewal. The dataset includes features like Customer Lifetime Value, Coverage, Education, Gender, Income, Monthly Premium Auto, Number of Policies, and encoded fields such as Sales Channel and Vehicle Class.

Output:
Python code that preprocesses the dataset, cleans the data (e.g., handling missing values and ensuring numerical inputs), and then trains a classification model (e.g., logistic regression or decision tree). Ensure scaling is applied to features using StandardScaler, set `max_iter` in LogisticRegression to 500, evaluate on a test set, and print key performance metrics such as accuracy.
"""
````

## Call LLM to generate the Python code

````python
model_id = "anthropic.claude-3-sonnet-20240229-v1:0"

prompt = PROMPT_code.format(INPUT_CONTENT=text, 
                            FUNCTION_DESCRIPTION=FUNCTION_DESCRIPTION)

messages = [{"role": "user", "content": [{"type": "text", "text": prompt}]}]

output_obj = \
invoke_claude_base(client, 
                   messages,
                   system = None,
                   model_id=model_id, 
                   temperature = 0.5, 
                   top_k = None, 
                   top_p = None,
                   stop_sequences=["Human:"],
                   use_streaming = True,
                   anthropic_version = "bedrock-2023-05-31",
                   print_details = True)
````

## Parse the Python code from the LLM response

````python
import re
response_text = output_obj['response_text']
find_list = [x.start() for x in re.finditer('```',response_text)]
code_start = find_list[0]+3+len('python\n')
code_end = find_list[1]
generated_code = response_text[code_start: code_end]
````

````python
from pygments import highlight
from pygments.lexers import PythonLexer
from pygments.formatters import TerminalFormatter

print(highlight(generated_code, PythonLexer(), TerminalFormatter()))
````

````python
# save the code to a file
with open('model_code.py', 'w') as f:
    f.write(generated_code)
````
